## Pricelist Generator v. 2.0

Read the Excel file

```python
import pandas as pd
import numpy as np
```

```python
pricelist_full = pd.ExcelFile('Pricelist_2022.xlsx')
```

Read all the sheet names in the Excel file, send it to a variable "sheets" and print it.

```python
sheets = pricelist_full.sheet_names
print(sheets)
```

Each sheet will be send to a Data frame

Process all the sheets to a Data frame

```python
for csheet in sheets[2:-4]:

    print("Processing sheet: " + csheet)

    #Read the Sheet and convert it in a DataFrame
    current_sheet = pricelist_full.parse(csheet, skiprows=6)
    
    #Order the Data Frame a bit:
    current_sheet.rename(columns = {"Pos.": "position",
                    "Order number": "product_code",
                    "Amount": "amount",
                    "Description": "description",
                    "Single price": "unit_price_no_tax",
                    "Total": "total_no_tax",
                    "Discount": "discount",
                    "Discount.1": "discount_no_tax",
                    "Final price": "final_price_no_tax",
                    "Unnamed: 9": "final_price",
                    "Unnamed: 10": "unit_price"},
         inplace=True)

    current_sheet.head()

    #Add Column with the name of the machine
    current_sheet["System"] = csheet

    """Extraemos la moneda en que estan los precios,
        teniendo en cuenta que en la fila 0, columna
        Single price se encuentra este valor
        Asi mismo se añande a una nueva columna
    """
    currency = current_sheet["unit_price_no_tax"].iloc[0]
    #Changing the index of the column by a number
    current_sheet["Currency"] = currency

    print(current_sheet.head())

    test_result = current_sheet[current_sheet["product_code"].isnull() == False][["product_code", "description", "unit_price_no_tax","System","Currency"]]

    #with pd.ExcelWriter ("test_"+csheet+".xls") as writer:
    #    current_sheet(writer)

    test_result.to_excel("MetrologyPl"+ csheet +".xlsx", sheet_name="Sheet1"+ csheet)
```
